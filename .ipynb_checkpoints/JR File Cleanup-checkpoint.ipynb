## Analysis
- Narrowed down to my columns
- Replaced Nans in my column

```python
import pandas as pd
import numpy as np

file = "./indeed_job_dataset.csv"

original_df = pd.read_csv(file)
original_df.head()
```

```python
#Reduce columns to mine only
columns = [
    "Location",
    "Company_Revenue",
    "python",
    "sql",
    "machine learning",
    "r",
    "hadoop"
]
#print(columns)
reduced_columns_df = original_df[columns]
reduced_columns_df
```

```python
#Clean Company Revenue
reduced_columns_df['Company_Revenue'] = reduced_columns_df['Company_Revenue'].replace(np.nan, "Missing")
reduced_columns_df
```

```python
#Spot test columns
comp_rev_test = reduced_columns_df['Company_Revenue'].unique()
print(comp_rev_test)
loc_test = reduced_columns_df['Location'].unique()
print(loc_test)
loc_count_test= reduced_columns_df['Location'].value_counts()
print(loc_count_test)
```

```python
#Write to csv for group merged file
reduced_columns_df.to_csv("JR_Cleaned_Df.csv",index=True)
```
