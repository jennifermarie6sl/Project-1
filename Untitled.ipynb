```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# Read Indeed Dataset and store into Pandas data frame
indeed_df = pd.read_csv("./indeed_job_dataset.csv")
indeed_df
```

```python
def search(text,n):
    '''Searches for text, and retrieves n words either side of the text, which are retuned seperatly'''
    word = r"\W*([\w]+)"
    groups = re.search(r'{}\W*{}{}'.format(word*n,'place',word*n), text).groups()
    return groups[:n],groups[n:]
```

```python
results = []

for value in indeed_df['Description']:
    t = value
    results.append(search(t, 20))
    
results
```
